## Defining feature set for the most "fun" song

I personally dislike heavy metal music, even if it is "fun", but some people like it. 

As "fun" is subjective and can be in every genre of music, the concept of genres should be dissacosiated with how fun the song is. This is so that the result can be as general as possible. Therefore, my philosophy of choosing the characteristics of a fun song is not based on my pre-conceived notion of what"fun" songs should be like. Rather, it is based on scientific evidence that some music characteristics evokes the necessary human emotions for a song to be considered "fun". 

With that in mind, I chose the following characteristics. Attached is a quick justification, after doing some research. 

Energy: High energy songs tend to be uplifting and a confidence booster. 

Mode: Songs played in a major key are judged as happy, while songs in minor key are judged as sad. (2)

Valence: Based on the data description given: "Tracks with high valence sound more positive (e.g. happy, cheerful, euphoric), while tracks with low valence sound more negative (e.g. sad, depressed, angry)". 

Tempo: "Stressing, Expressive and Amusing are the words gotten from a high tempo, whilst the opposite terms Relaxing, Expressionless and Boring are obtained for a low tempo." (1)

Other characteristics seem to be associated with genres (danceability, accousticness, loudness, speechiness). Thus, their omission is necessary so that the result is not biased towards a specific genre. 


Sources: 
(1) https://www.ncbi.nlm.nih.gov/pmc/articles/PMC4971092/
(2) https://www.nme.com/blogs/nme-blogs/the-science-of-music-why-do-songs-in-a-minor-key-sound-sad-760215

```python
spotify_new = spotify[spotify.columns[[0,1,2,4,7,12,13]]]
spotify_new.head()
```

```python
spotify_new.describe()
```

```python
# To get rid of all songs played in minor key. 
spotify_new = spotify_new[spotify_new["mode"] != 0.0]
```

```python
# To give equal weightage to the rest of the characteristics, each characteristics is divided into 5 bins. 
# Energy and valence have values between 0 to 1. Thus, it is easy to divide it into bins. 
spotify_new.loc[:,"energy"] = pd.cut(spotify_new["energy"], bins=[0, 0.2, 0.4, 0.6, 0.8,1], include_lowest=True, labels=[1,2,3,4,5]).values.astype('int64')
spotify_new.loc[:,"valence"] = pd.cut(spotify_new["valence"], bins=[0, 0.2, 0.4, 0.6, 0.8,1], include_lowest=True, labels=[1,2,3,4,5]).values.astype('int64')
# Tempo has values between 64 to 199. To divide into 5 bins, each bin would have an interval of 27. 
spotify_new.loc[:,"tempo"] = pd.cut(spotify_new["tempo"], bins=[64,91,118,145,172,199], include_lowest=True, labels=[1,2,3,4,5]).values.astype('int64')
```

```python
spotify_new.head()
```

```python
# Score adds up the bin number of energy, valence and tempo for each row. The row with the highest score can be said
# to be the most fun song. 
spotify_new["score"] = spotify_new["energy"].add(spotify_new["valence"]).add(spotify_new["tempo"])
spotify_new.loc[spotify_new["score"].idxmax()]
```

## Result

According to the aobve procedure, Dejala que vuelva by Piso 21 appears to be the most fun song of 2018. It scored highly in terms of energy, valence and tempo.
